```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

uploaded = files.upload()

df = pd.read_csv('Day_12_banking_data.csv')

total_transaction_amount = df.groupby('Account_Type')['Transaction_Amount'].sum()
total_transaction_amount.plot(kind='bar', color='skyblue')
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

transactions_per_branch = df['Branch'].value_counts()
transactions_per_branch.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')
plt.show()
```
